```python
target = 'Income'

mean_baseline = [df2017tract[target].mean()] * len(df2017tract)
df2017tract['Mean Baseline'] = df2017tract[target].mean()
df2017tract['Error'] = df2017tract['Mean Baseline'] - df2017tract[target]
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
```

```python
df2017tract.columns
```

```python
# Running basic linear regression model, excluding all direct income related columns
# Also excluding string values

features = ['TotalPop', 'Men', 'Women', 'Hispanic',
       'White', 'Black', 'Native', 'Asian', 'Pacific', 'VotingAgeCitizen',
       'Poverty',
       'ChildPoverty', 'Professional', 'Service', 'Office', 'Construction',
       'Production', 'Drive', 'Carpool', 'Transit', 'Walk', 'OtherTransp',
       'WorkAtHome', 'MeanCommute', 'Employed', 'PrivateWork', 'PublicWork',
       'SelfEmployed', 'FamilyWork', 'Unemployment']
target = 'Income'

X = df2017tract[features]
y = df2017tract[target]
```

```python
model = LinearRegression(SimpleImputer(strategy='median'))
model.fit(X, y)
```

```python
model.coef_, model.intercept_
```

```python
y_pred = model.predict(X)
```

```python
mean_absolute_error(y, y_pred)
```

Improved MAE from 22,836 to 11,246 from mean baseline with Linear Regression model

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df2017tract, test_size=0.2)
```

```python
train.shape, test.shape
```

```python
target = 'Income'
y_train = train[target]
y_test = test[target]

features = ['State', 'County', 'TotalPop', 'Men', 'Women', 'Hispanic',
       'White', 'Black', 'Native', 'Asian', 'Pacific', 'VotingAgeCitizen',
        'Poverty',
       'ChildPoverty', 'Professional', 'Service', 'Office', 'Construction',
       'Production', 'Drive', 'Carpool', 'Transit', 'Walk', 'OtherTransp',
       'WorkAtHome', 'MeanCommute', 'Employed', 'PrivateWork', 'PublicWork',
       'SelfEmployed', 'FamilyWork', 'Unemployment']
X_train = train[features]
X_test = test[features]
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_log_error

def rmsle(y_true, y_pred):
    return np.sqrt(mean_squared_log_error(y_true, y_pred))
```

```python
y_pred = np.full_like(y_test, fill_value=y_train.mean())
print('Validation RMSLE, Mean Baseline:', rmsle(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
import category_encoders as ce
from sklearn.pipeline import make_pipeline

pipeline = make_pipeline(ce.OrdinalEncoder(),
                         RandomForestRegressor(n_estimators=100, n_jobs=-1))
pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)
print('RMSLE with Random Forest: ', rmsle(y_test, y_pred))
```

```python
rf = pipeline.named_steps['randomforestregressor']
importances = pd.Series(rf.feature_importances_, X_train.columns)
plt.figure(figsize=(8,12))
importances.sort_values().plot.barh(color='orange');
```

```python
plt.figure(figsize=(9, 6))
sns.distplot(y_test, label='True')
sns.distplot(y_pred, label='Predicted')
plt.legend();
```
